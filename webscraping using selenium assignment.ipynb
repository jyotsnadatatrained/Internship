```python
#questionQ1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
#have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
#jobs data.
#This task will be done in following steps:
#1. First get the webpage https://www.shine.com/
#2. Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter the location” field.
#3. Then click the searchbutton.
#4. Then scrape the data for the first 10 jobs results you get.
#5. Finally create a dataframe of the scraped data.
#Note: All of the above steps have to be done in code. No step is to be done manually.

#get the webpage

driver.get('https://www.shine.com/')
```

```python
#search for the criteria

designation = driver.find_element(By.CLASS_NAME,'form-control  ')
designation.send_keys('Data Analyst')
```

```python
#search for the criteria

location = driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input')
location.send_keys('Bangalore')
```

```python
#click on the search criteria

search = driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/div[2]/div[2]/div/form/div/div[2]/div/button')
search.click()
```

```python
#create empty lists

job_titles=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
#append all empty list

title_tags=driver.find_elements(By.XPATH,'//h2[@itemprop="name"]')
for i in title_tags[0:10]:
    title=i.text
    job_titles.append(title)
    
location_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in location_tags[0:10]:
    location=i.text
    job_location.append(location)

company_tags=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')
for i in company_tags[0:10]:
    company=i.text
    company_name.append(company) 
    
experience_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in experience_tags[0:10]:
    exp=i.text
    experience_required.append(exp)
                                
```

```python
print(len(job_titles),len(job_location),len(company_name),len(experience_required))
```

```python
#create dataframe

import pandas as pd
df = pd.DataFrame({'Title':job_titles,'Location':job_location,'company':company_name,'Experience':experience_required})
df
```

```python
#append the empty list

for row in rows[1:]:
    columns = row.find_elements(By.TAG_NAME,'td')
    name = columns[0].text
    born_dead = columns[1].text
    term = columns[2].text
    remarks = columns[3].text
    prime_ministers.append({"Name": name, "Born-Dead": born_dead, "Term of office": term, "Remarks": remarks})
```

```python
#create dataframe

df = pd.DataFrame(prime_ministers)
```
