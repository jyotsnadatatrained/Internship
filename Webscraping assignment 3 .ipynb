```python
#question1Write a python program which searches all the product under a particular product from www.amazon.in. The 
#product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for 
#guitars. 


# Import the necessary libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

# Function to search for a product on Amazon
def search_product(product):
    # Set up the Chrome driver
    driver = webdriver.Chrome()
    
    # Navigate to the Amazon website
    driver.get("https://www.amazon.in/")
    
    try:
        # Find the search input field and enter the product name
        search_input = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "twotabsearchtextbox")))
        search_input.send_keys(product)
        
        # Submit the search form
        search_input.submit()
        
        # Wait for the search results to load
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "s-result-list")))
        
        # Print the search results
        search_results = driver.find_elements(By.CSS_SELECTOR, ".s-result-item .a-link-normal")
        for result in search_results:
            print(result.text)
        
    except TimeoutException:
        print("Timeout occurred while searching for the product.")
        
    

# Get the product name from the user
product_name = input("Enter the product name: ")


search_product(product_name)
```

```python
#question3 Write a python program to access the search bar and search button on images.google.com and scrape 10 
#images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.

import time
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains

# Function to scrape images for a given keyword
def scrape_images_for_keyword(keyword, num_images=10):
    # Create a WebDriver instance (e.g., Chrome WebDriver)
    driver = webdriver.Chrome()

    try:
        # Navigate to Google Images
        driver.get("https://www.google.com/imghp")

        # Find the search bar element and enter the keyword
        search_box = driver.find_element(By.NAME, "q")
        search_box.send_keys(keyword)

        # Submit the search
        search_box.send_keys(Keys.RETURN)

        # Scroll down to load more images (you may need to adjust the scroll count)
        for _ in range(3):
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)

        # Find image elements
        image_elements = driver.find_elements(By.XPATH, "//img[@class='rg_i Q4LuWd']")

        # Create a directory to save the images
        if not os.path.exists(keyword):
            os.makedirs(keyword)

        # Download the images
        for i, img_element in enumerate(image_elements[:num_images], start=1):
            img_url = img_element.get_attribute("src")
            if img_url and not img_url.startswith("data:image"):
                img_extension = img_url.split(".")[-1]
                img_filename = f"{keyword}/{keyword}_{i}.{img_extension}"

                # Right-click on the image to open the context menu and select "Save image as..."
                ActionChains(driver).context_click(img_element).send_keys(Keys.ARROW_DOWN).send_keys(Keys.RETURN).perform()
                time.sleep(2)

                # Rename the downloaded file
                os.rename("image.png", img_filename)

                print(f"Downloaded image {i}/{num_images} for '{keyword}'")

    except Exception as e:
        print(f"An error occurred: {str(e)}")

```

```python
#4 Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com
#and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand 
#Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, 
#“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the 
#details is missing then replace it by “- “. Save your results in a dataframe and CSV. 

import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

def scrape_flipkart_smartphones(keyword):
    # Create a WebDriver instance (e.g., Chrome WebDriver)
    driver = webdriver.Chrome()

    try:
        # Navigate to Flipkart
        driver.get("https://www.flipkart.com/")

        # Close the login popup if it appears
        try:
            driver.find_element(By.CLASS_NAME, "_2KpZ6l._2doB4z").click()
        except Exception:
            pass

        # Find the search bar element and enter the keyword
        search_box = driver.find_element(By.NAME, "q")
        search_box.send_keys(keyword)
        search_box.send_keys(Keys.RETURN)

        # Wait for the page to load (you may need to adjust the wait time)
        driver.implicitly_wait(10)

        # Initialize data lists
        brand_names = []
        smartphone_names = []
        colors = []
        ram_list = []
        storage_list = []
        primary_cameras = []
        secondary_cameras = []
        display_sizes = []
        battery_capacities = []
        prices = []
        product_urls = []

        # Find product containers
        product_containers = driver.find_elements(By.CLASS_NAME, "_4ddWXP")

        for product in product_containers:
            # Extract details for each product
            brand_name = product.find_element(By.CLASS_NAME, "_4rR01T").text
            smartphone_name = product.find_element(By.CLASS_NAME, "IRpwTa").text
            details = product.find_elements(By.CLASS_NAME, "IRpwTa")
            color = details[0].text if len(details) > 0 else "-"
            ram = details[1].text if len(details) > 1 else "-"
            storage = details[2].text if len(details) > 2 else "-"
            primary_camera = details[3].text if len(details) > 3 else "-"
            secondary_camera = details[4].text if len(details) > 4 else "-"
            display_size = details[5].text if len(details) > 5 else "-"
            battery_capacity = details[6].text if len(details) > 6 else "-"
            price = product.find_element(By.CLASS_NAME, "_30jeq3").text
            product_url = product.find_element(By.TAG_NAME, "a").get_attribute("href")

            # Append details to the lists
            brand_names.append(brand_name)
            smartphone_names.append(smartphone_name)
            colors.append(color)
            ram_list.append(ram)
            storage_list.append(storage)
            primary_cameras.append(primary_camera)
            secondary_cameras.append(secondary_camera)
            display_sizes.append(display_size)
            battery_capacities.append(battery_capacity)
            prices.append(price)
            product_urls.append(product_url)

        # Create a DataFrame
        data = {
            "Brand Name": brand_names,
            "Smartphone Name": smartphone_names,
            "Colour": colors,
            "RAM": ram_list,
            "Storage(ROM)": storage_list,
            "Primary Camera": primary_cameras,
            "Secondary Camera": secondary_cameras,
            "Display Size": display_sizes,
            "Battery Capacity": battery_capacities,
            "Price": prices,
            "Product URL": product_urls,
        }

        df = pd.DataFrame(data)

        # Save to CSV
        df.to_csv("flipkart_smartphones.csv", index=False)

        print("Data scraped and saved to 'flipkart_smartphones.csv'")

    except Exception as e:
        print(f"An error occurred: {str(e)}")

    finally:
        # Close the WebDriver
        driver.quit()


```
